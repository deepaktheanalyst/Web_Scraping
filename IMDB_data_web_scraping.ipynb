```python
import csv
header = ['Data']
file_name = "movies.csv"
with open(file_name, 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    for serial_no in range(1,11288,50):
        link = (f"https://www.imdb.com/search/title/?title_type=tv_series,tv_miniseries&genres=action&start={serial_no}&ref_=adv_prv")
        data = r.get(link).text
        soup = bs(data)
        for row in soup.find_all('div',attrs={'class':'lister-item-content'}):
            movie_data = row.text.replace('\n','').strip()
            writer.writerow([movie_data])
```

```python
with open('IMDB_data.csv', mode='w', newline='',encoding='utf-8') as IMDB_file:
    IMDB_writer = csv.writer(IMDB_file)
    IMDB_writer.writerow(['title', 'certificate','duration','genre' 'rating', 'stars', 'votes'])

    for serial_no in range(1,11288,50):
        link = (f"https://www.imdb.com/search/title/?title_type=tv_series,tv_miniseries&genres=action&start={serial_no}&ref_=adv_prv")
        data = r.get(link).text
        soup = bs(data)
        divs = soup.find_all('div', attrs={'class':'lister-item-content'})
        for i in divs:
            try:
                title = i.find('h3', attrs={'class':'lister-item-header'}).text.replace('\n',"")
            except:
                title = "N/A"
            try:
                certificate = i.find('span',attrs={'class':'certificate'}).text
            except:
                certificate = "N/A"
            try:
                duration = i.find('span',attrs={'class':'runtime'}).text
            except:
                duration = "N/A"
            try:
                genre = i.find('span',attrs={'class':'genre'}).text.replace('\n','')
            except:
                genre = "N/A"    
            try:
                rating = i.find('div', attrs={'class':'inline-block ratings-imdb-rating'}).text.replace('\n', '')
            except:
                rating = "N/A"
            try:
                stars = i.find('p', attrs={'class':''}).text.replace('\n', '').split('|')
            except:
                stars = "N/A"
            try:
                votes = i.find('span', attrs={'name':'nv'}).text
            except:
                votes = "N/A"
            IMDB_writer.writerow([title,certificate,duration,genre,rating, stars, votes])
```
