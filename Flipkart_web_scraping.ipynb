```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
r = requests.get("https://www.flipkart.com/search?q=Broen%20shoes&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
```

```python
soup = bs(r.text,'html.parser')
```

```python
results = soup.find_all('div',attrs={'class':'_1xHGtK _373qXS'})
len(results)
```

```python
records = []
for result in results:
    brand_name = result.find('div',attrs= {'class':'_2WkVRV'}).text
    shoes_name = result.find('a',attrs= {'class':'IRpwTa'}).text
    price = result.find('div',attrs={'class':'_30jeq3'}).text
    url = result.find('a',attrs={'class':'IRpwTa'})['href']
    image = result.find('img',attrs={'class':'_2r_T1I'})['src']
    records.append((brand_name, shoes_name, price,url,image))
```

```python
import pandas as pd  
df = pd.DataFrame(records, columns=['Brand_Name', 'Shoes_name', 'Price','url','image_url']) 
```
